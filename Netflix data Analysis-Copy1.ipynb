# Agar percentage ke basis par karna hai jaise ki 25%,50%,75% or 100% ke basis par uske liye ham ek function ka use karte hai jisse kaam easy ho jayega

```python
def categorize_column(df,col,label):
    edges=[
        df[col].describe()['min'],
        df[col].describe()['25%'],
        df[col].describe()['50%'],
        df[col].describe()['75%'],
        df[col].describe()['max']
    ]

    df[col]=pd.cut(df[col],edges,labels=label,duplicates='drop')
    return df
```

```python
label=["Not_Popular","below_average","Average","Popular"]

categorize_column(df,'Vote_Average',label)
```
